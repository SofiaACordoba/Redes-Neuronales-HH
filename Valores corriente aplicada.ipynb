## Veamos distintos valores de Iapp (corriente aplicada) para determinar un rango de la misma

Recordar que las condiciones iniciales son: [-65, 0, 0, -65, 0, 0] 

```python
import matplotlib.pyplot as plt
import numpy as np
from neurona import neurona
```

**0,1 <= Iapp <= 2.3.**  Hemos notado que para la corriente aplicada entre esos valores, el potencial se termina 'apagando'

```python
red0 = neurona()
red0.Iapp = 0.5
t0, Y0 = red0.simular_red() 
red0.graficar_potenciales_de_membrana_a_lo_largo_del_tiempo(t0,Y0)

red1 = neurona()
red1.Iapp = 2.3
t, Y = red1.simular_red() 
red1.graficar_potenciales_de_membrana_a_lo_largo_del_tiempo(t,Y)
```

**2.4 <= Iapp <= 2.8**    Se puede decir que 'gana' V1 o V2

```python
red2 = neurona()
red2.Iapp = 2.4
t2, Y2 = red2.simular_red() 
red2.graficar_potenciales_de_membrana_a_lo_largo_del_tiempo(t2,Y2)

red3 = neurona()
red3.Iapp = 2.5
t3, Y3 = red3.simular_red() 
red3.graficar_potenciales_de_membrana_a_lo_largo_del_tiempo(t3,Y3)

red4 = neurona()
red4.Iapp = 2.6
t4, Y4 = red4.simular_red() 
red4.graficar_potenciales_de_membrana_a_lo_largo_del_tiempo(t4,Y4)

red5 = neurona()
red5.Iapp = 2.7
t5, Y5 = red5.simular_red() 
red5.graficar_potenciales_de_membrana_a_lo_largo_del_tiempo(t5,Y5)

red6 = neurona()
red6.Iapp = 2.8
t6, Y6 = red6.simular_red() 
red6.graficar_potenciales_de_membrana_a_lo_largo_del_tiempo(t6,Y6)
```

**2.9 <= Iapp <= 3.1**    Se observan distintos momentos en que una queda inhibida por otra.

```python
valores_Iapp = [2.9, 3.0, 3.1]  

for i, Iapp in enumerate(valores_Iapp):
    red = neurona()
    red.Iapp = Iapp
    t, Y = red.simular_red()
    red.graficar_potenciales_de_membrana_a_lo_largo_del_tiempo(t, Y)

```
