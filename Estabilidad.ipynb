```python
from neurona import neurona
import matplotlib.pyplot as plt
import numpy as np
```

```python
red = neurona()
iapp = np.arange(2, 3.6, 0.001)
autovalores_parte_real = []
autovalores_parte_imaginaria = []

for i in iapp:
    y_star, J, eigvals, eigvecs = red.analisis_estabilidad(i)
    
    autovalores_parte_real.append(np.max(eigvals.real))
    autovalores_parte_imaginaria.append(np.max(eigvals.imag))

# Gráficos
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(iapp, autovalores_parte_real, '-b', linewidth=2)
plt.axhline(0, color='k', linestyle='--') 
plt.xlabel("Iapp")
plt.ylabel("Parte real máxima")
plt.title("Max Re(λ) vs Iapp")

plt.subplot(1, 2, 2)
plt.plot(iapp, autovalores_parte_imaginaria, '-r', linewidth=2)
plt.axhline(0, color='k', linestyle='--') 
plt.xlabel("Iapp")
plt.ylabel("Parte imaginaria máxima")
plt.title("Max Im(λ) vs Iapp")

plt.tight_layout()
plt.show()
```
